# 02 — Feature Engineering
Extract ~100+ heat-related features per site × year combination.

**Input**: DB1 weather data  
**Output**: `data/feature_matrix.csv`

```python
import pandas as pd
import os, sys

# Navigate to repo root
if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('..')
sys.path.insert(0, '.')

from src.features import identify_heatwaves, build_feature_matrix
print('Imports ready | Working dir:', os.getcwd())
```

```python
# ── Load weather data ──
weather = pd.read_excel('data/DB1_PRISM_Weather_Daily_1981_2023.xlsx', sheet_name='Data')
print(f'Weather: {weather.shape}')
print(f'Columns: {weather.columns.tolist()}')
```

```python
# ── Quick test of heatwave detection ──
test = [35, 36, 39, 40, 38.5, 37, 35, 39, 41, 39, 36]
hws = identify_heatwaves(test)
print(f'Test data: {test}')
print(f'Heatwaves found: {len(hws)}')
for hw in hws:
    print(f'  Days {hw["start"]}–{hw["end"]}: {hw["duration"]}d, HWU={hw["hwu"]:.1f}°C')
```

```python
# ── Build feature matrix (takes ~30 seconds) ──
feature_matrix = build_feature_matrix(weather)
display(feature_matrix.head())
```

```python
# ── Sanity checks ──
feat_cols = [c for c in feature_matrix.columns if c not in ['site', 'year']]
print(f'Site×year combos: {len(feature_matrix)}  (paper: ~213)')
print(f'Features: {len(feat_cols)}')
print(f'Sites: {sorted(feature_matrix["site"].unique())}')
print(f'Years: {feature_matrix["year"].min()} to {feature_matrix["year"].max()}')
print(f'\nAvg season heat days: {feature_matrix["season_heat_days"].mean():.2f}')
print(f'Max season heat days: {feature_matrix["season_heat_days"].max()}')
print(f'Avg season heatwaves: {feature_matrix["season_no_hw"].mean():.2f}')
```

```python
# ── Save for next notebook ──
feature_matrix.to_csv('data/feature_matrix.csv', index=False)
print(f'Saved: data/feature_matrix.csv ({feature_matrix.shape})')
```
