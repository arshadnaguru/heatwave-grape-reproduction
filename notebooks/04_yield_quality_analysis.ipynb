## Meeting Prep — Key Points

**Methodology understanding:**
- HCA classifies seasons by heat pattern → 3 types (Cool, PRE-V, POST-V)
- RF validates clusters are real and stable (~1.35% error)
- Attribution: compare yield/quality across season types within same vineyards
- Mixed models account for vineyard-level random effects

**Why these results matter:**
- First empirical quantification of heatwave damage at commercial scale
- 30% yield loss from preveraison heat is massive for industry
- Differential effects on fruit quality → winemaking implications

**What you'd do differently / extend:**
- Use supervised ML (XGBoost, LSTM) for yield *prediction* rather than classification
- Incorporate phenology-based features (FL→VER interval)
- Add management practices as covariates
- Scale to more varieties and regions using transfer learning

**Key quote for the meeting:**
> "This paper uses classical unsupervised methods to classify seasons. The natural AI extension is *prediction* — using the same weather features with supervised ML to forecast yield before harvest, enabling proactive mitigation."
