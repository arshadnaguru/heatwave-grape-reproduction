Load and inspect all 4 datasets from Previtali et al. (2026).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# If running from notebooks/ folder, go to repo root
if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('..')
print('Working dir:', os.getcwd())
```

```python
# ── Load all datasets ──
weather = pd.read_excel('data/DB1_PRISM_Weather_Daily_1981_2023.xlsx', sheet_name='Data')
phenology = pd.read_excel('data/DB2_Phenology.xlsx', sheet_name='Data')
yield_data = pd.read_excel('data/DB3_Yield_Harvest_date.xlsx', sheet_name='Data')
fruit_comp = pd.read_excel('data/DB4_Fruit_composition.xlsx', sheet_name='Data')

print('All data loaded!')
print(f'  Weather:    {weather.shape}')
print(f'  Phenology:  {phenology.shape}')
print(f'  Yield:      {yield_data.shape}')
print(f'  Fruit comp: {fruit_comp.shape}')
```

```python
# ── Weather: DB1 ──
print('WEATHER — Columns:', weather.columns.tolist())
weather['date'] = pd.to_datetime(weather['date'])
print(f'Date range: {weather["date"].min()} to {weather["date"].max()}')
print(f'Sites: {sorted(weather["Site"].unique())}')
display(weather.head())
display(weather.describe().round(2))
```

```python
# ── Phenology: DB2 ──
print('PHENOLOGY — Columns:', phenology.columns.tolist())
print(f'Seasons: {phenology["Season"].min()} to {phenology["Season"].max()}')
display(phenology.head())
```

```python
# ── Yield: DB3 ──
print('YIELD — Columns:', yield_data.columns.tolist())
print(f'Seasons: {yield_data["Season"].min()} to {yield_data["Season"].max()}')
print(f'Blocks: {yield_data["Block_ID"].nunique()}')
display(yield_data.head())
display(yield_data.describe().round(2))
```

```python
# ── Fruit Composition: DB4 ──
print('FRUIT COMP — Columns:', fruit_comp.columns.tolist())
print(f'Seasons: {fruit_comp["Season"].min()} to {fruit_comp["Season"].max()}')
display(fruit_comp.head())
display(fruit_comp.describe().round(2))
```

```python
# ── Quick viz: Tmax over time for one site ──
site1 = weather[weather['Site'] == weather['Site'].unique()[0]].copy()
site1 = site1[(site1['date'].dt.month >= 4) & (site1['date'].dt.month <= 10)]

fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(site1['date'], site1['tmax'], lw=0.3, alpha=0.7)
ax.axhline(38, color='red', ls='--', label='38°C threshold')
ax.set(xlabel='Date', ylabel='Tmax (°C)',
       title=f'Daily Tmax — {site1["Site"].iloc[0]} (growing season only)')
ax.legend()
plt.tight_layout()
plt.show()
print(f'Days above 38°C: {(site1["tmax"] >= 38).sum()} out of {len(site1)}')
```

### Summary
- Weather data is clean, daily, 5 sites, 1981–2023
- No `Season` column in weather — extract year from `date`
- Yield data uses `Season` (= year), `Block_ID`, `Avg_Harvest_DOY`, `Yield_tha`
- Fruit composition: 12 analytes, 2017–2023
